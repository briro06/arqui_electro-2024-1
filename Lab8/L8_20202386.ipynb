i) **(1.5p)** Cree 10 archivos de texto y 10 archivos binarios de distintos tamaños. Los tamaños para c/archivo (en $MB$) corresponden a potencias de 2 (de $2^4$ a $2^{12}$)



---
Con respecto al contenido de los archivos:


1.   Archivos de texto (extensión: `.txt`):




*   Utilize un bucle iterativo de tal manera de que en cada iteración se escriba un nuevo carácter en el archivo. Los caracteres deben ser letras mayúsculas del alfabeto inglés (`A` a `Z`).
*   Nombre de archivo: `text_file_{size}KB.txt` (ejm. *text_file_4KB.txt*)

2.   Archivos binarios (extensión: `.bin`):



*   ` os.urandom(n)` es una función del módulo `os` que retorna una cadena de $N$ caracteres aleatorios ($N\; bytes$), lo cual es de utilidad para aplicaciones de criptografía.
*   Nombre de archivo: `text_file_{size}KB.bin` (ejm. *binary_file_4KB.bin*)

```python
import os
import random
from random import randint
#Generamos una función para nuestros archivos  .txt
def generate_text_file(file_path, size_kb):
  #Vamos a escribir el archivo
  with open(file_path,'w') as f:
    #Realizamos un bucle que genera los archivos con su respectivo peso en KB
    for i in range((2**10)*size_kb):
      #Utilizamos la función randint para que nos pueda generar un carácter aleatorio ya que en a
      lt=randint(65,91)
      #Guardamos en una variable las letras de la A a la Z
      letr=chr(lt)
      f.write(letr)
  pass

#Generamos una función para nuestros archivos .bin
def generate_binary_file(file_path, size_kb):
  #Vamos a escribir el archivo
  with open(file_path,'wb') as f:
    #Generamos archivos
    letr=os.urandom((2**10)*size_kb)
    f.write(letr)
  pass

#Nuestro for tendrá que ir desde 4 hasta 14 , para que puedan estar los numeros 3,4,5,..13 e irá iterando y generando archivos con sus respectivos pesos
for i in range(4,14):
  peso_text=2**(i)
  #Creamos la ruta para los archivos .txt
  ruta_archivo1=f"text_file_{peso_text}KB.txt"
  #Creamos la ruta para los archivos .bin
  ruta_archivo2=f"text_file_{peso_text}KB.bin"

  #Hacemos llamado a la función que genera archivos .txt
  generate_text_file(ruta_archivo1,peso_text)
  #Hacemos llamado a la función que genera archivos .bin
  generate_binary_file(ruta_archivo2,peso_text)
```

ii) **(1.5p)** Es posible implementar distintos métodos de compresión en Python. Para ello, son de utilidad los módulos `gzip` y `bz2`. Cree 4 funciones, tal que cada una implemente las siguientes rutinas:

*   `compress_file_gzip`: Compresión de archivos utilizando `gzip`
*   `decompress_file_gzip`: Descompresión de archivos utilizando `gzip`
*   `compress_file_bz2`: Compresión de archivos utilizando `bz2`
*   `decompress_file_bz2`: Descompresión de archivos utilizando `bz2`

Para cada una de las funciones considere:

*   **Input:**

1.   `input_file` (nombre del archivo de entrada)
2.   `output_file` (nombre del archivo de salida)

*   **Output:**

1.   `exec_time` (tiempo de ejecución de la rutina. Utilice `perf_counter()`)



```python
from time import perf_counter
import gzip
import bz2
import os

#Comprimimos los archivos con gzip
def compress_file_gzip(input_file,output_file):
  #Creamos la variable que mide el inicio del tiempo
  i1=perf_counter()
  #Hacemos la lectura del archivo para luego proceder a comprimirlo con gzip
  with open(input_file,'rb') as f_in:
    with gzip.open(output_file,'wb') as f_out:
      f_out.writelines(f_in)
  #Usamos la variable f1 para medir el fin del tiempo 
  f1=perf_counter()
  tf1=(f1-i1)*1e6
  #Finalmente retornamos el tiempo que tomo en microsegundos
  return tf1

#Descomprimimos los archivos
def decompress_file_gzip(input_file,output_file):
  #Creamos la variable que mide el inicio del tiempo
  i2=perf_counter()
  #Hacemos la lectura del archivo para luego proceder a descomprimirlo con gzip
  with open(input_file,'rb') as f_in:
    with gzip.open(output_file,'wb') as f_out:
      f_out.writelines(f_in)
  #Medimos el fin del tiempo que le tomo descomprimir
  f2=perf_counter()
  tf2=(f2-i2)*1e6
  #Finalmente retornamos el tiempo que tomo en microsegundos
  return tf2
#Comprimimos los archivos
def compress_file_bz2(input_file,output_file):
  #Creamos la variable que mide el inicio del tiempo 
  i3=perf_counter()
  #Hacemos la lectura del archivo para luego proceder a comprimirlo con bz2
  with open(input_file,'rb') as f_in:
    with bz2.open(output_file,'wb') as f_out:
      f_out.writelines(f_in)
  #Medimos el fin del tiempo que le tomo descomprimir
  f3=perf_counter()
  tf3=(f3-i3)*1e6
  #Finalmente retornamos el tiempo que tomo en microsegundos
  return tf3

#Descomprimimos los archivos
def decompress_file_bz2(input_file,output_file):
  #Creamos la variable que mide el inicio del tiempo
  i4=perf_counter()
  #Hacemos la lectura del archivo para luego proceder a descomprimirlo con bz2
  with open(input_file,'rb') as f_in:
    with bz2.open(output_file,'wb') as f_out:
      f_out.writelines(f_in)
  #Medimos el fin del tiempo que le tomo descomprimir
  f4=perf_counter()
  tf4=(f4-i4)*1e6
  #Finalmente retornamos el tiempo que tomo en microsegundos
  return tf4

for i in range(4,14):
  #Calcularemos el peso del archivo para cada iteración
  peso_text=2**(i)
  #Archivos de entrada
  input_file1=f"text_file_{peso_text}KB.txt"
  input_file2=f"text_file_{peso_text}KB.bin"

  #Archivos de salida
  output_file1=f"text_file_{peso_text}_txt_KB.gz" #entrada txt -> gzip
  output_file2=f"text_file_{peso_text}_bin_KB.bz2"#entrada bin  -> bz2
  output_file3=f"text_file_{peso_text}_bin_KB.gzip" #entrada bin -> gzip
  output_file4=f"text_file_{peso_text}_txt_KB.bz2" #entrada txt -> bz2

  #Para los archivos txt usando gzip
  compress_file_gzip(input_file1,output_file1)
  decompress_file_gzip(output_file1,f'out_descompress_gzip_{peso_text}.txt')

  #Para los archivos bin usando gzip
  compress_file_gzip(input_file2,output_file3)
  decompress_file_gzip(output_file3,f'out_descompress_gzip_{peso_text}.bin')

  #Para los archivos txt utilizando bz2
  compress_file_bz2(input_file2,output_file4)
  decompress_file_bz2(output_file4,f'out_descompress_bz2_{peso_text}.txt')

  #Para los archivos bin utilizando bz2
  compress_file_bz2(input_file2,output_file2)
  decompress_file_bz2(output_file2,f'out_descompress_bz2_{peso_text}.bin')
```
